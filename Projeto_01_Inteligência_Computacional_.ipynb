```python
import numpy as np

from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 100
dados = np.random.rand(N)

plt.figure(figsize=(6,4))
plt.title('Histograma')
plt.hist(dados, bins=10)
plt.grid();
```

```python
dados_originais = 18 + (102 * np.random.rand(N)).astype(int)

plt.figure(figsize=(6,4))
plt.title('Idades')
plt.hist(dados_originais, bins=10)
plt.grid();
```

```python
dados_normalizados = (dados_originais - np.min(dados_originais)) / (np.max(dados_originais) - np.min(dados_originais))

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title('Idades normalizadas')
plt.hist(dados_normalizados, bins=10)
plt.grid()
plt.subplot(1,2,2)
plt.title('Idades originais')
plt.hist(dados_originais, bins=10)
plt.grid();
```

```python
dados_normalizados = (dados_originais - np.mean(dados_originais)) / np.std(dados_originais, ddof=1)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title('Idades normalizadas')
plt.hist(dados_normalizados, bins=10)
plt.grid()
plt.subplot(1,2,2)
plt.title('Idades originais')
plt.hist(dados_originais, bins=10)
plt.grid();
```

```python
N = 100
x = np.zeros((N,2))
x[:,0] = np.linspace(0,1,N) + (.08*np.random.randn(N))
x[:,1] = 5 + 3*np.linspace(0,1,N) + (.08*np.random.randn(N))
y = np.linspace(0,1,N) > .5

plt.figure(figsize=(6,4))
plt.title('Dados originais')
plt.scatter(x[:,0], x[:,1])
plt.grid();
```

```python
pca = PCA(n_components=2)

x_pca = pca.fit_transform(x)

plt.figure(figsize=(6,4))
plt.title('Dados com PCA')
plt.scatter(x_pca[:,0], x_pca[:,1])
plt.grid();
```

```python
N = 100
x = np.zeros((N,2))
x[:,0] = np.linspace(0,1,N) + (.08*np.random.randn(N))
x[:,1] = 5 + 3*np.linspace(0,1,N) + (.08*np.random.randn(N))
y = np.linspace(0,1,N) > .5

plt.figure(figsize=(6,4))
plt.title('Dados com classes')
plt.scatter(x[:,0], x[:,1], c=y)
plt.grid();
```

```python
modelo = Pipeline([('pca', PCA(n_components=2)),
                   ('clf', LogisticRegression())])

modelo.fit(x, y)
modelo.score(x, y)
```
